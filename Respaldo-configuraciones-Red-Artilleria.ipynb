# <font color = '990000'> Respaldar los archivos NETWORK y WIRELESS de la Red Artilleria <font>

```python
import os, serial, time
ser=serial.Serial(port='com12', baudrate=115200)
```

## <font color = 'blue'> Ejemplo de Comandos Linux via serial <font>

### <font color = 'blue'> (Vaciar el Buffer) <font>

```python
print(ser.isOpen())
ser.close()
```

```python
ser.open()
```

```python
ser.write(b"ls" + b'\r\n')
ser.write(b"ls /etc"+ b'\r\n')

out = ''
time.sleep(1)
while ser.inWaiting() > 0:
    out += ser.read(1).decode()
if out != '':
    print (out)
```

## <font color = 'blue'> Respaldar en PC archivos WIRELESS, NETWORK y FIREWALL <font>

### <font color = 'green'> Network </font>

```python
ser.write(b"cat /etc/config/network" + b'\r\n')
out = ''
time.sleep(1)
while ser.inWaiting() > 0:
    out += ser.read(1).decode()
if out != '':
    print (out)
```

```python
dispositivo=out.splitlines()[-1][5:-4]
text="\n".join(out.splitlines()[1:-2])
print(text)
with open ("Respaldo configuraciones Artilleria/network "+dispositivo, "w", newline="") as f: #Crea un archivo con la informacion de /etc/config/network
    f.write ('#  '+out.splitlines()[-1]+out.splitlines()[0])
    f.write (text)
```

```python
with open ("Respaldo configuraciones Artilleria/network "+dispositivo, "r") as f: #lee el archivo network
    print(f.read())
```

## <font color = 'FF3300'> Respaldar en radio consolidado de archivos WIRELESS, NETWORK Y FIREWALL </font>

```python
ser.write(b'mkdir /etc/config/"Respaldo Red Artilleria"'+b'\n')
ser.write(b'cd /etc/config/"Respaldo Red Artilleria"'+b'\n')
```

```python
for elemento in os.listdir("Respaldo configuraciones Artilleria/"):
    with open ("Respaldo configuraciones Artilleria/" + elemento, "r") as f:
        contenido = f.read()
    elemento=elemento.replace(" ", "-")
    ser.write(b'cat > ' + bytes(elemento, 'utf-8') + b'\n')
    time.sleep(.1)
    ser.write(bytes(contenido, 'utf-8')+b'\n')
    time.sleep(1)
    ser.write(b'\x03')
    out = ''
    time.sleep(1)
    while ser.inWaiting() > 0:
        try:
            out += ser.read(1).decode()
        except e:
            print(e)
    if out != '':
        print (out)
```
